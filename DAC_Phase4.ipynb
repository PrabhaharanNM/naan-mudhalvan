```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('country_vaccinations.csv')


print(df.info())


print(df.describe())




plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='total_vaccinations', data=df, marker='o')
plt.title('Total Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.show()


plt.figure(figsize=(14, 8))
sns.barplot(x='iso_code', y='daily_vaccinations', data=df)
plt.title('Daily Vaccinations by Country')
plt.xlabel('Country ISO Code')
plt.ylabel('Daily Vaccinations')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(14, 8))
sns.boxplot(x='iso_code', y='total_vaccinations_per_hundred', data=df)
plt.title('Total Vaccinations per Hundred by Country')
plt.xlabel('Country ISO Code')
plt.ylabel('Total Vaccinations per Hundred')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(10, 8))
sns.scatterplot(x='total_vaccinations', y='people_vaccinated', data=df)
plt.title('Total Vaccinations vs. Total People Vaccinated')
plt.xlabel('Total Vaccinations')
plt.ylabel('Total People Vaccinated')
plt.show()




correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
